```python
file = open("text_file.txt", "r")
```

```python
text = file.read()
```

```python
word_counts = {}
```

```python
for word in text.split():
    if word in word_counts:
        word_counts[word] += 1
    else:
        word_counts[word] = 1 
```

```python
print(f'{"WORD":<12}COUNT')

for word, count in sorted(word_counts.items()):
    print(f'{word:<12}{count}')

print('\nNumber of unique words:', len(word_counts))
```
